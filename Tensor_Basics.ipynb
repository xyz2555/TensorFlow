```python
import tensorflow as tf
# Mengimpor library TensorFlow dan memberinya alias 'tf'
```

```python
x = tf.constant([[1, 2, 3], [4, 5, 6]])
# Membuat tensor konstan 2D (matrix 2x3)

print(x)
# Menampilkan isi tensor x ke layar
```

```python
x = tf.eye(4)
# Membuat matriks identitas 4x4 (diagonal bernilai 1, lainnya 0)

print(x)
# Menampilkan matriks identitas
```

```python
x = tf.random.normal((3,3), mean=0, stddev=1)
# Membuat tensor acak berukuran 3x3 dengan distribusi normal
# mean = 0 dan standar deviasi = 1

print(x)
# Menampilkan tensor acak
```

```python
x = tf.range(2)
# Membuat tensor 1D berisi nilai [0, 1]

print(x)
# Menampilkan tensor range

x = tf.cast(x, dtype=tf.float32)
# Mengubah tipe data tensor dari integer ke float32

print(x)
# Menampilkan tensor setelah casting
```

```python
x = tf.constant([1, 2, 3])
# Membuat tensor 1D berisi vektor [1, 2, 3]

y = tf.constant([4, 5, 6])
# Membuat tensor 1D berisi vektor [4, 5, 6]

z = tf.tensordot(x, y, axes=1)
# Melakukan operasi dot product antara x dan y
# axes=1 artinya menjumlahkan hasil perkalian elemen yang bersesuaian

print(z)
# Menampilkan hasil dot product (skalar)

print(tf.add(x, y))
# Menjumlahkan tensor x dan y secara elemen-per-elemen
```

```python
x = tf.random.normal((2,2))
# Membuat tensor acak 2x2

print(x)
# Menampilkan tensor TensorFlow

x = x.numpy()
# Mengonversi tensor TensorFlow menjadi NumPy array

print(x)
# Menampilkan NumPy array

x = tf.convert_to_tensor(x)
# Mengonversi kembali NumPy array menjadi TensorFlow tensor
```
